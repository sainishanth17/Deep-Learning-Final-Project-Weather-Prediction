## Read data

```python
csv_path = "../Data/jena_climate_2009_2016.csv"
df = pd.read_csv(csv_path, index_col=0, parse_dates=True)
```

```python
df.head()
```

I choose the first 20000 timestamps as sub-data and used 720 as lag. The model checks out the 720 timestamps and uses them for future predictions. (this selection is based on keras website example for time series modeling). later I apply the model for 72 timestamps prediction ahead (12 hours ahead).

```python
column_name = 'T (degC)'
start = 0
end = 20000

training_df = df[column_name][start:end]
```

## Autoregressive (AR)


Here I use different lags value and see how it improve the AR model estimation. Later we make a visualization for four different lags.

```python
def fit_ar_at_lags(lag_value, ahead_value, show_past, i):
    
    # train autoregression
    model = AutoReg(training_df, lags=lag_value)
    ar_model_fit = model.fit()
    
    
    # use start and end for sub_df
    start = len(training_df)
    ahead = ahead_value
    end = start + ahead
    test = df['T (degC)'][start:end]
    
    
    
    # make predictions
    predictions = ar_model_fit.predict(start=len(training_df), 
                                    end=len(training_df)+len(test)-1, 
                                    dynamic=False)

    # apply same index for prediction df
    predictions.index = test.index
    
    return predictions, test, start, ahead, end
    
```

```python
def plot_ar_at_algs(lag_value_list, ahead_value, show_past):

    # plot 
    plt.figure(figsize=(20,12))
    
    
    for i, lag_value in enumerate(lag_value_list):
        
    
        # fit and pred
        predictions, test, start, ahead, end = fit_ar_at_lags(lag_value, ahead_value, show_past, i+1)

        # plot
        plt.subplot(2,2,i+1)
        plt.plot(predictions, label=f'AR prediction lag = {lag_value}')
        plt.plot(test, 'r^', label='Data - Expected')

        past = show_past

        plt.xlabel('Date')
        plt.ylabel('Temperature c')
        plt.title("Temperature prediction")

        star = df['T (degC)'][start:start+1].index
        plt.axvline(star,color='k', linestyle='--', label='test_line')
        #     print(start-past)
        #     print(start)
        plt.plot(df[column_name][(start-past):(start)], c='r')

        plt.ylim(0,)

        plt.legend()
    
```

```python
lag_value_list = [10,72, 720, 2000]
plot_ar_at_algs(lag_value_list, ahead_value=200, show_past=100)
```

But using the above code is not efficient. It's better to train the ar models once and call it for prediction. 
Above code fit the ar model every single time that is called for ploting.

```python
def train_ar_for_lag(lag_value):
    # train autoregression
    model = AutoReg(training_df, lags=lag_value)
    ar_model_fit = model.fit()
    return ar_model_fit
```

```python
lag_value_list = [10,72,720,2000]

ar_lg_10 = train_ar_for_lag(lag_value=10)
ar_lg_72 = train_ar_for_lag(lag_value=72)
ar_lg_720 = train_ar_for_lag(lag_value=720)
ar_lg_2000 = train_ar_for_lag(lag_value=2000)
```

```python
model_list = [ar_lg_10, ar_lg_72,
              ar_lg_720, ar_lg_2000]
```

```python
def plot_ar_at_algs(lag_value_list, ahead_value, show_past, model_list):

    # plot 
    plt.figure(figsize=(20,12))
    
    
    for i, lag_value in enumerate(lag_value_list):
        
    
        # fit and pred
        # use start and end for sub_df
        start = len(training_df)
        ahead = ahead_value
        end = start + ahead
        test = df['T (degC)'][start:end]

        # make predictions
        predictions = model_list[i].predict(start=len(training_df), 
                                        end=len(training_df)+len(test)-1, 
                                        dynamic=False)

        # apply same index for prediction df
        predictions.index = test.index
        

        # plot
        plt.subplot(2,2,i+1)
        plt.plot(predictions, label=f'AR prediction lag = {lag_value}')
        plt.plot(test, 'r^', label='Data - Expected')

        past = show_past

        plt.xlabel('Date')
        plt.ylabel('Temperature c')
        plt.title("Temperature prediction")

        star = df['T (degC)'][start:start+1].index
        plt.axvline(star,color='k', linestyle='--', label='test_line')
        #     print(start-past)
        #     print(start)
        plt.plot(df[column_name][(start-past):(start)], c='r')

        plt.ylim(0,)

        plt.legend()
    
```

```python
plot_ar_at_algs(lag_value_list, ahead_value=20, show_past=100, model_list=model_list)
```

let's make a gif visualization

```python
@gif.frame
def plott(i):
    
    plot_ar_at_algs(lag_value_list, ahead_value=i, show_past=100, model_list=model_list)

frames = []    
for i in tqdm(range(1,720)):

    
    frame =  plott(i)
    frames.append(frame)
```
